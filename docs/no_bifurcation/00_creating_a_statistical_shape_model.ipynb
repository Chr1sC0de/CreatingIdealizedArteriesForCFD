# Creating a Statistical Shape Model
We can create a statistical shape model using the CylindricalPointModel class in the pymethods library

```python
import pymethods as pma
import pyvista as pv

model = pma.algorithms.shapeModelling.CylindricalPointModel(
        r'./base_models', structured_shape=(3, 360, 720))
model.build(scale=True, modes=10)
print("the fraction of explained variance is:", model.explained_variance())
```

We can now generate a new model by modifying the value of the modes

```python
shape = model.generateShape([0, -1, 0])
mesh  = pv.StructuredGrid(*shape)
mesh.plot()
```

```python
shape = model.generateShape([3,  -1, -2, -2, 1, 2,-1,3,3,2,-3,3])
mesh = pv.StructuredGrid(*shape)
mesh.plot()
```

Note: as the shape model will vary depending on the base shape when making models for deep learning it is best to build multiple models and sample from all the datasets
